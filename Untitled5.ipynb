```python
import pandas as pd 
import kagglehub
import os 
import matplotlib.pyplot as plt
```

```python
path = kagglehub.dataset_download("abdallahwagih/company-employees")
files = os.listdir(path)
excel_files = [f for f in files if f.endswith(".xlsx")]
if excel_files:
    excel_path = os.path.join(path, excel_files[0])
    df = pd.read_excel(excel_path)
    print("First 5 rows:")
    print(df.head())
else:
    print("No Excel files found in the folder.")
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
for col in ["Gender", "Department", "Country", "Center"]:
    print(f"\n--- Value counts for '{col}' ---")
    print(df[col].value_counts())
```

```python
# Pie Chart for Gender
df["Gender"].value_counts().sort_values(ascending=False).plot(kind="pie", autopct='%1.1f%%', startangle=90)
plt.title("Gender Distribution")
plt.ylabel("")  # Hide y-axis label
plt.show()
```

```python
# Bar Chart for Center
df["Center"].value_counts().sort_values(ascending=False).plot(kind="bar", color='skyblue')
plt.title("Employee Count per Center")
plt.xlabel("Center")
plt.ylabel("Number of Employees")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Histogram for Job Rate
df["Job Rate"].plot(kind="hist", bins=10, color="orange", edgecolor="black")
plt.title("Distribution of Job Rate")
plt.xlabel("Job Rate")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
df["Job Rate"].describe()
```

```python
df.groupby("Department")["Annual Salary"] \
    .mean() \
    .sort_values(ascending=False) \
    .head(7) \
    .plot(kind="bar", color="teal", edgecolor="black")

plt.title("Average Salary Graph by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df.groupby("Center")["Monthly Salary"] \
    .mean() \
    .sort_values(ascending=False) \
    .plot(kind="bar", color="salmon", edgecolor="black")

plt.title("Average Monthly Salary by Center")
plt.xlabel("Center")
plt.ylabel("Average Monthly Salary")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df.groupby("Country")["Job Rate"] \
    .mean() \
    .sort_values(ascending=False) \
    .plot(kind="bar", color="lightgreen", edgecolor="black")

plt.title("Job Rates by Countries")
plt.xlabel("Country")
plt.ylabel("Average Job Rate")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.hist(df["Overtime Hours"], bins=10, color='steelblue', edgecolor='black')
plt.title("Histogram of Overtime Hours")
plt.xlabel("Overtime Hours")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
X = df[["Years", "Job Rate"]]       
y = df["Annual Salary"]   
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
len(y_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
predslr = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(predslr,y_test)
```

```python
import joblib
joblib.dump(lr,"linearmodel.pkl")
```
